```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# linspace returns an array of evenly spaced numbers as specified from start to stop
n = 1000
d = 50
T = 1
t = np.linspace(0, T, n) 
dt = t[1] - t[0]
dB = np.sqrt(dt) * np.random.normal(size=(n-1, d))    # Bt2 - Bt1 ~ N(0, t2-t1)
B0 = np.zeros(shape=(1, d))
B = np.concatenate((B0, np.cumsum(dB, axis=0)) , axis = 0)
```

```python
plt.plot(t, B)
plt.show()
```

```python
def quadratic_variation(B):
    return np.cumsum(np.power(np.diff(B, axis = 0, prepend=0.), 2), axis=0)
```

```python
plt.plot(t, quadratic_variation(B)) # QV of Brownian motion is T
plt.show()
```
